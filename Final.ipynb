# Problem 1 : Login to your Instagram

```python
def LogIn(username, password):
    
    try :
        
        #locating username textbox and sending username
        user_name = wait.until(EC.presence_of_element_located((By.NAME,'username')))
        user_name.send_keys(username)
        
        #locating password box and sending password
        pwd = driver.find_element_by_name('password')
        pwd.send_keys(password)
        
        #locating login button 
        login_button = driver.find_element(By.CSS_SELECTOR, 'button[type="submit"]')
        login_button.click()
        
        #Save Your Login Info? : Not Now
        pop = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="_ac8f"]')))
        pop.click()
        
        time.sleep(5)
        
        #Turn on notificaton : Not Now
        button = wait.until(EC.presence_of_element_located((By.XPATH,'//button[@class="_a9-- _a9_1"]')))
        button.click()
        
    except TimeoutException :
        print ("Something went wrong! Try Again")
```

# Problem 2 : Type for “food” in search bar

```python
def search(s):
    try:
        
        #locating and clicking on search icon to make search visible
        search = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@aria-label="Search"]')))
        search.click()
        
        #locating serch bar and sending text
        search_box = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aauy')))
        search_box.send_keys(s)
        
        #waiting till all searched is located
        wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1uhb9sk x1plvlek xryxfnj x1c4vz4f x2lah0s x1q0g3np xqjyukv x6s0dn4 x1oa3qoh x1nhvcw1"]/span')))
        
        time.sleep(5)
        
        #extracting all serched handle
        handle_names = driver.find_elements_by_xpath('//*[@class="x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1uhb9sk x1plvlek xryxfnj x1c4vz4f x2lah0s x1q0g3np xqjyukv x6s0dn4 x1oa3qoh x1nhvcw1"]/span')
        names = []
        
        #extracting username
        for i in handle_names :
            names.append(i.get_attribute('innerHTML'))

        time.sleep(5)
        #clearing search bar
        driver.find_element_by_xpath('//div[@aria-label="Clear the search box"]').click()
        
        time.sleep(7)
        
        #locating and clicking on search icon to close search 
        search = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@aria-label="Search"]')))
        search.click()
        
        return names   
    
    except TimeoutException :
        print('No Search Found!')
```

```python
#extracting all the names of the Instagram Handles that are displayed in list after typing “food” using search('food')
driver.find_element_by_xpath('//a[@href = "/"]').click() #To return to the home screen
name_list = search('food')
for i in name_list :
    print(i)
```

# Problem 4 : Follow/Unfollow given handle

```python
def follow():
    try :
        
        #locating the follow button
        follow_button = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1i64zmx x1n2onr6 x1plvlek xryxfnj x1iyjqo2 x2lwn1j xeuugli xdt5ytf xqjyukv x1qjc9v5 x1oa3qoh x1nhvcw1"]')))
        
        # Check if already following
        if follow_button.text == 'Follow' : 
            follow_button.click()
            time.sleep(3)
            print("Following the page")
        
        else : 
            print('Already Following')

    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def unfollow():
    try :
        
        #locating the follow button
        follow_button = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1i64zmx x1n2onr6 x1plvlek xryxfnj x1iyjqo2 x2lwn1j xeuugli xdt5ytf xqjyukv x1qjc9v5 x1oa3qoh x1nhvcw1"]')))
        
        #checking for text
        if follow_button.text !='Follow' : 
            follow_button.click()
            time.sleep(2)
            #locating popup window (when you click on follow button)
            wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class = "x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1uhb9sk x1plvlek xryxfnj x1iyjqo2 x2lwn1j xeuugli xdt5ytf xqjyukv x1cy8zhl x1oa3qoh x1nhvcw1"][last()]')))
            pop_up = driver.find_elements_by_xpath('//div[@class = "x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1uhb9sk x1plvlek xryxfnj x1iyjqo2 x2lwn1j xeuugli xdt5ytf xqjyukv x1cy8zhl x1oa3qoh x1nhvcw1"]')
            
            #location the unfollow option the pop up window
            for i in pop_up :
                if i.text == 'Unfollow' :
                    i.click()
                    break
            
            print("Unfollowed the page")
                    
            time.sleep(3)
            
        else : 
            print('Already Unfollowed')

    except TimeoutException :
        print("Something Went Wrong!")
```

```python
#for following this instgram handle
follow()
#for unfollowing this instgram handle
unfollow()
```

# Problem 5 : Like/Unlike posts

```python
def Like_Post():
    try :
        #scrolling for locating post
        driver.execute_script('window.scrollTo(0, 6000);')
        time.sleep(3)
        driver.execute_script('window.scrollTo(0, -6000);')
        time.sleep(3)
        #locating post
        posts = driver.find_elements_by_class_name('_aagw')
     
        for i in range(30):
            posts[i].click()
            time.sleep(2)
            #locating like/unke button
            like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "_aamw"]/button')))
            st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            
            #checking if we have already liked this post
            if st == 'Like' :
                like.click()
                time.sleep(2)
            else :
                print('You have already LIKED Post Number', i+1)
                time.sleep(2)
            #locating cross button for closing post
            a = driver.find_element_by_xpath('//div[@class = "x1i10hfl x6umtig x1b1mbwd xaqea5y xav7gou x9f619 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz x6s0dn4 x78zum5 xjbqb8w x1ejq31n xd10rxx x1sy0etr x17r0tee x1ypdohk xl56j7k x1y1aw1k x1sxyh0 xwib8y2 xurb0ha"]')
            a.click()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def Unlike_Post():
    try :
        #scrolling for locating post
        driver.execute_script('window.scrollTo(0, 6000);')
        time.sleep(3)
        driver.execute_script('window.scrollTo(0, -6000);')
        time.sleep(3)
        #locating post
        posts = driver.find_elements_by_class_name('_aagw')
     
        for i in range(30):
            posts[i].click()
            time.sleep(2)
            #locating like/unke button
            like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "_aamw"]/button')))
            st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            
            #checking if we have already unliked this post
            if st == 'Unlike' :
                like.click()
                time.sleep(2)
            else :
                print('You have already UNLIKED Post Number', i+1)
                time.sleep(2)
            #locating cross button for closing post
            a = driver.find_element_by_xpath('//div[@class = "x1i10hfl x6umtig x1b1mbwd xaqea5y xav7gou x9f619 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz x6s0dn4 x78zum5 xjbqb8w x1ejq31n xd10rxx x1sy0etr x17r0tee x1ypdohk xl56j7k x1y1aw1k x1sxyh0 xwib8y2 xurb0ha"]')
            a.click()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def Following():
    try :
        # locating following button and click on it
        following_button = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class = "x1i10hfl xjbqb8w x6umtig x1b1mbwd xaqea5y xav7gou x9f619 x1ypdohk xt0psk2 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz _alvs _a6hd"]')))
        following_button[1].click()
        
        time.sleep(5)
        
        #locating followers list
        frame = driver.find_element_by_class_name('_aano')
        #scrolling untill all users are located
        for i in range(20):
            time.sleep(1)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",frame)
        
        time.sleep(2)

        names = []

        #extracting userdata
        following = driver.find_elements_by_xpath('//span[@class = "_aacl _aaco _aacw _aacx _aad7 _aade"]')
        
        time.sleep(5)

        #extracting username
        for i in following :
            names.append(i.text)
            
        time.sleep(5)


        #location close button for follower window    
        close_button = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//button[@class = "_abl-"]')))
        close_button[1].click()

        return names
    
    except TimeoutException :
        print("Something Went Wrong!")
```
